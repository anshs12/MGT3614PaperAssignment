```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Data
years = ['2020', '2021', '2022', '2023']

# DraftKings
draftkings_assets = [404771673, 4069054000, 4040152000, 3944866000]
draftkings_liabilities = [742540000, 2390526000, 2717459000, 3104560000]

# MGM
mgm_assets = [36494934000, 40899116000, 45692206000, 42368548000]
mgm_liabilities = [25250000000, 29770000000, 40320000000, 38000000000]

# Caesars
caesars_assets = [36385000, 38031000, 33527000, 33366000]
caesars_liabilities = [31351000, 33490000, 29776000, 28646000]

# Plotting
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Function to format y-axis ticks
def millions_or_billions(x, pos):
    'The two args are the value and tick position'
    if x >= 1_000_000_000:
        return f'{x / 1_000_000_000:.0f}B'
    elif x >= 1_000_000:
        return f'{x / 1_000_000:.0f}M'
    else:
        return f'{x:.0f}'

formatter = FuncFormatter(millions_or_billions)

# DraftKings
axs[0, 0].bar(years, draftkings_assets, color='skyblue', label='Assets')
axs[0, 0].bar(years, draftkings_liabilities, color='orange', label='Liabilities', bottom=draftkings_assets)
axs[0, 0].set_title('DraftKings')
axs[0, 0].set_ylabel('Amount')
axs[0, 0].legend()
axs[0, 0].yaxis.set_major_formatter(formatter)

# MGM
axs[0, 1].bar(years, mgm_assets, color='skyblue', label='Assets')
axs[0, 1].bar(years, mgm_liabilities, color='orange', label='Liabilities', bottom=mgm_assets)
axs[0, 1].set_title('MGM')
axs[0, 1].legend()
axs[0, 1].yaxis.set_major_formatter(formatter)

# Caesars
axs[0, 2].bar(years, caesars_assets, color='skyblue', label='Assets')
axs[0, 2].bar(years, caesars_liabilities, color='orange', label='Liabilities', bottom=caesars_assets)
axs[0, 2].set_title('Caesars')
axs[0, 2].legend()
axs[0, 2].yaxis.set_major_formatter(formatter)

# Only Liabilities
axs[1, 0].bar(years, draftkings_liabilities, color='orange', label='DraftKings')
axs[1, 1].bar(years, mgm_liabilities, color='orange', label='MGM')
axs[1, 2].bar(years, caesars_liabilities, color='orange', label='Caesars')

# Adding labels to the second row
axs[1, 0].set_title('Liabilities Only')
axs[1, 0].legend()

for ax in axs.flat:
    ax.set_xlabel('Year')

plt.tight_layout()
plt.show()
```
